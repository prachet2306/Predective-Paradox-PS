# Demand Prediction Model

**Objective:** Build a 1-hour-ahead forecasting model for electricity demand using historical grid data, weather conditions, and macroeconomic indicators.

This notebook covers:
1. Exploratory Data Analysis (EDA) & Preprocessing on all the 3 files
2. Advanced Feature Engineering (Cyclical encoding, Rolling EWM, Lag features)
3. Training and testing using XGBoost

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb
from sklearn.metrics import mean_absolute_percentage_error

sns.set_theme(style="whitegrid")
import warnings
warnings.filterwarnings('ignore')
```

## 1. Power Data: Loading & Inspection
First, loading the historical power demand dataset, inspecting its shape, and sorting the dates.

```python
# Reading the file:
df_power = pd.read_excel('PGCB_date_power_demand.xlsx') # Adjust filename if needed

# Checking the shape and columns:
print(f"Power Data Shape: {df_power.shape}")
display(df_power.head(20))
```

As we can see the datetimes are not sorted, so sort them.

```python
# Sorting and then resetting the index
df_power = df_power.sort_values('datetime').reset_index(drop=True)
```

### 1.1 Selecting required columns
Only datetime, demand, and remarks columns are required, since we only need to know exactly when the power was used, how much was used, everything else is just distracting background noise that doesn't help us predict the future.

```python
# Selecting required columns
df_power = df_power[['datetime' , 'demand_mw']]

# Renaming columns for better readability
df_power = df_power.rename(columns = { 'datetime' : 'DateTime' ,  
                                       'demand_mw' : 'Demand (in MW)' 
                                        })
```

### 1.2 Resampling & Structuring the Time Axis 
Looking at the first 20 rows, it is pretty clear that the time stamps are not all at a difference of 1 hour, there are some time stamps having 30 minute gaps, however only 1 reading per hour is required.

```python
# The resample function takes the max demand for that hour and adds NaN values in hour timestamps which are not present in the original file
df_hourly = df_power.resample('1h', on='DateTime').max().reset_index()
```

### 1.3 Missing Values Analysis
Analyzing the missing values: 

```python
# Check for missing values
df_power.isna().sum()
```

As visible, there are 1051 missing values,but we cannot directly fill in all missing vales, we need to check for continuous gaps because safely bridging a tiny, one-hour glitch is easy, but trying to mathematically invent days of lost history is just making things up.

```python
# Code for finding continuous gaps in the data:
missing_mask = df_power['Demand (in MW)'].isnull()
missing_groups = (missing_mask != missing_mask.shift()).cumsum()
contiguous_gaps = missing_mask.groupby(missing_groups).sum()

print(f"\nLargest contiguous gap of missing demand data: {contiguous_gaps.max()} hours")
```

### 1.4 Imputation
Since there 517 hours of continuous missing data (almost a week), we cannot simply linearly interpolate, so seasonal imputation is also used:

```python
# Filling missing hours using the exact same time and day from recent weeks (168 hrs = 1 week)
for weeks_back in [1, 2, 3, 4]:
    df_hourly['Demand (in MW)'] = df_hourly['Demand (in MW)'].fillna(
        df_hourly['Demand (in MW)'].shift(168 * weeks_back)
    )
df_hourly['Demand (in MW)'] = df_hourly['Demand (in MW)'].interpolate(method='linear')
df_hourly['Demand (in MW)'] = df_hourly['Demand (in MW)'].bfill()
```

```python
# Checking if NaNs are left
df_power.isna().sum()
```

### 1.5 Cyclical Encoding
Perfoming cyclical encoding on the datafarame give us new column: hour_sin, hour_cos, day_sin, day_cos, month_sin, month_cos

```python
def preprocess_power_data(df):
    df = df.copy()
    
    df["hour"] = df['DateTime'].dt.hour
    df["dayofweek"] = df['DateTime'].dt.dayofweek
    df["month"] = df['DateTime'].dt.month

    # Cyclic encoding
    df["hour_sin"] = np.sin(2 * np.pi * df["hour"] / 24)
    df["hour_cos"] = np.cos(2 * np.pi * df["hour"] / 24)

    df["day_sin"] = np.sin(2 * np.pi * df["dayofweek"] / 7)
    df["day_cos"] = np.cos(2 * np.pi * df["dayofweek"] / 7)

    df["month_sin"] = np.sin(2 * np.pi * df["month"] / 12)
    df["month_cos"] = np.cos(2 * np.pi * df["month"] / 12)

    df.drop(columns=["hour", "dayofweek", "month"], inplace=True)
    return df
df_cyclic = preprocess_power_data(df_hourly)
```

```python
# Check the new columns!
df_cyclic.head()
```

### 1.6 Outlier detection and fixing
We will use IQR method for outlier detection. We calculate a custom "normal operating range" for every specific hour of the day. If a power reading falls violently outside its designated hourly window, the code flags it as a broken sensor and mathematically smooths over the gap using the surrounding data. We fix the outliers using linear interpolation.

```python

from pyod.models.iforest import IForest


# We use Hour and Month sin/cos so the model understands daily and seasonal regimes
features = ['Demand (in MW)', 'hour_sin', 'hour_cos', 'month_sin', 'month_cos']

# Make sure there are no NaNs in these columns before feeding to PyOD
X = df_cyclic[features].copy()

# contamination = 0.005 means we are telling the model we expect roughly 0.5% of the data to be broken sensors
clf_iforest = IForest(contamination=0.005, random_state=42)

# Train the model and get predictions (0 = inlier, 1 = outlier)
clf_iforest.fit(X)
outlier_labels = clf_iforest.labels_

# We convert 1s (outliers) to True for our anomaly mask
anomaly_mask_pyod = (outlier_labels == 1)

# We don't want PyOD deleting tiny 50 MW fluctuations at night just because they look mathematically "isolated"
rolling_median = df_hourly['Demand (in MW)'].rolling(window=24, center=True, min_periods=1).median()
deviation = np.abs(df_hourly['Demand (in MW)'] - rolling_median)

final_anomaly_mask = anomaly_mask_pyod & (deviation > 500)

# 6. Destroy and Impute
df_cyclic.loc[final_anomaly_mask, 'Demand (in MW)'] = np.nan
df_cyclic['Demand (in MW)'] = df_hourly['Demand (in MW)'].interpolate(method='spline', order=3)

print(f"PyOD Isolation Forest detected {final_anomaly_mask.sum()} severe, out-of-context spikes.")
```

Let's check the fixed outliers in a particular month of a year using graph

```python
import time

print("Starting clean up...")
start_time = time.time()

df_cyclic['Demand_Cleaned'] = df_cyclic['Demand (in MW)'].copy()

# Apply mask and use FAST linear interpolation
df_cyclic.loc[final_anomaly_mask, 'Demand_Cleaned'] = np.nan
df_cyclic['Demand_Cleaned'] = df_cyclic['Demand_Cleaned'].interpolate(method='linear')

print(f"Interpolation finished in {round(time.time() - start_time, 2)} seconds!")

# Zoom in on a specific month (e.g., October 2023)
mask_zoom = (df_cyclic['DateTime'].dt.year == 2017) & (df_cyclic['DateTime'].dt.month == 5)
df_zoom = df_cyclic[mask_zoom]

# Build the Before/After Plot
plt.figure(figsize=(16, 6))

# Plot BEFORE (Raw Data) in transparent Red
plt.plot(df_zoom['DateTime'], df_zoom['Demand (in MW)'], 
         color='#d62728', alpha=0.7, label='Before (Raw with Anomalies)', linewidth=1.5)

# Plot AFTER (Cleaned Data) in solid Blue
plt.plot(df_zoom['DateTime'], df_zoom['Demand_Cleaned'], 
         color='#1f77b4', alpha=0.9, label='After (Cleaned)', linewidth=2)

# Formatting
plt.title('Power Demand: Before vs. After Contextual Anomaly Correction (Oct 2023)', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Demand (MW)', fontsize=12)
plt.legend(loc='upper left', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()


plt.show()
```

## 3. Merge and add rolling statistics and lag features

## 3.1 Merge the 3 files!
This section performs a left join to align annual economic indicators with cyclical and weather data by shifting the year forward. It then extracts temporal features like the day of the week before cleaning up the helper merge columns.

```python
df_macro_shifted = df_economic_final.copy()
df_macro_shifted['Year'] = df_macro_shifted['Year'] + 1

df_master = pd.merge(df_cyclic, df_weather, on='DateTime', how='inner')


df_master['Year'] = df_master['DateTime'].dt.year


df_master = pd.merge(df_master, df_macro_shifted, on='Year', how='left')

df_master['DayOfWeek'] = df_master['DateTime'].dt.dayofweek

df_master = df_master.drop(columns=['Year'])


df_master.head()

```

The dataframe is merged!

### 3.2 Add Lag features and Is_weekend column
This step creates a binary flag to distinguish weekends from weekdays and introduces time-lagged demand features to capture historical trends. Shifting the data by 1 and 24 hours helps the model learn from both immediate past behavior and daily patterns.

```python
df_master['Is_Weekend'] = np.where(df_master['DayOfWeek'] >= 5, 1, 0)

df_master['Demand_Lag_1h'] = df_master['Demand (MW)'].shift(1)
df_master['Demand_Lag_24h'] = df_master['Demand (MW)'].shift(24)
```

### 3.3 Add rolling statistics and Target column
This section calculates exponentially weighted moving averages for demand and temperature to smooth out short-term noise while highlighting recent trends. It also defines the target variable by shifting demand back one hour, setting up the prediction objective.

```python
df_master['Demand_ewm_24h'] = df_master['Demand (MW)'].ewm(span=24, adjust=False).mean()
df_master['Temp_ewm_24h'] = df_master['Apparent_Temp_C'].ewm(span=24, adjust=False).mean()

df_master['Target_Next_Hour'] = df_master['Demand (MW)'].shift(-1)

df_master = df_master.dropna().reset_index(drop=True)
```

## 4. Training the model
We decided to use the XGBoost model to train the data.

```python
train_mask = df_master['DateTime'].dt.year <= 2023
test_mask = df_master['DateTime'].dt.year == 2024


X = df_master.drop(columns=['DateTime', 'Target_Next_Hour'])
y = df_master['Target_Next_Hour']

X_train, y_train = X[train_mask], y[train_mask]
X_test, y_test = X[test_mask], y[test_mask]

print(f"Training on {len(X_train)} hours (Up to 2023)")
print(f"Testing on {len(X_test)} hours (Exactly 2024)\n")


print("Training XGBoost Model")

model = xgb.XGBRegressor(
    n_estimators=500,        # Number of decision trees
    learning_rate=0.05,      # How fast the model learns
    max_depth=6,             # Maximum depth of each tree
    random_state=42,
    n_jobs=-1                # Uses all available CPU cores for speed
)


model.fit(X_train, y_train)

predictions = model.predict(X_test)

mape = mean_absolute_percentage_error(y_test, predictions)

print(f"2024 Deployment MAPE: {mape:.2%}")
```

### 4.1 Checking the feature importance
We extract the gain scores from our trained XGBoost model to identify which variables most significantly drive our energy demand predictions. We then visualize these rankings in a sorted bar plot to clearly see which features provide the most predictive value.

```python
importance_scores = model.feature_importances_

df_importance = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': importance_scores
})

df_importance = df_importance.sort_values(by='Importance', ascending=False).reset_index(drop=True)

plt.figure(figsize=(12, 8))

sns.barplot(x='Importance', y='Feature', data=df_importance, palette='viridis')

plt.title('XGBoost Feature Importance' , fontsize=16, fontweight='bold')
plt.xlabel('Relative Importance (Contribution to Accuracy)', fontsize=12)
plt.ylabel('Feature', fontsize=12)
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()

plt.show()
```

As we can see, the model depends too much on demand_mw, so there is a new strategy to fix this and obtain lower MAPE 

## 5. The Delta Strategy
We build short-term memory into our model so it understands exactly how power demand moves over time. By calculating the "deltas" (the mathematical difference between current hour and next hour), we capture the grid's physical momentum. This teaches the algorithm to sense whether demand is currently accelerating into a heavy peak or rapidly cooling off, giving it the context it needs to forecast the next hour accurately.

```python
df_master['Target_Delta'] = df_master['Target_Next_Hour'] - df_master['Demand (MW)']

X_delta = df_master.drop(columns=['DateTime', 'Target_Next_Hour', 'Target_Delta', 'Demand (MW)'])
y_delta = df_master['Target_Delta']

train_mask = df_master['DateTime'].dt.year <= 2023
test_mask = (df_master['DateTime'] >= '2024-01-01 00:00:00') & \
            (df_master['DateTime'] <= '2025-06-17 11:00:00')

X_train_delta, y_train_delta = X_delta[train_mask], y_delta[train_mask]
X_test_delta, y_test_delta = X_delta[test_mask], y_delta[test_mask]

current_demand_test = df_master.loc[test_mask, 'Demand (MW)']
actual_future_demand = df_master.loc[test_mask, 'Target_Next_Hour']

model_delta = xgb.XGBRegressor(n_estimators=500, learning_rate=0.05, max_depth=6, random_state=42, n_jobs=-1)
model_delta.fit(X_train_delta, y_train_delta)


predicted_deltas = model_delta.predict(X_test_delta)

final_predictions = current_demand_test + predicted_deltas

delta_mape = mean_absolute_percentage_error(actual_future_demand, final_predictions)
print(f"MAPE: {delta_mape:.4%}\n")


df_delta_imp = pd.DataFrame({
    'Feature': X_train_delta.columns,
    'Importance': model_delta.feature_importances_
}).sort_values(by='Importance', ascending=False).reset_index(drop=True)

plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=df_delta_imp, palette='mako', hue='Feature', legend=False)
plt.title('Delta Architecture', fontsize=16, fontweight='bold')
plt.xlabel('Relative Importance (Contribution to predicting the Delta)', fontsize=12)
plt.ylabel('Feature', fontsize=12)
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```

We can see that a macro indicator is covering the top spot, How?
Reason: it tackles our biggest mathematical problem first: the massive 10-year growth of the power grid. It acts as a heavy anchor to establish the base demand for the current era we are looking at, which perfectly sets the stage for our short-term momentum features to step in and fine-tune the exact hour-by-hour forecast. However, let's try to change this importance by hiding some features.

### 5.1 Test by hiding different features
We streamline our dataset by dropping redundant weather metrics like the smoothed temp_ewm. Instead, we strictly rely on apparent_temperature because the physical "feels like" condition—which factors in humidity and wind—is what actually drives humans to crank up their ACs or heaters.Also, removing the demand_ewm by the same logic. By removing these overlapping mathematical signals, we give our algorithm a cleaner, uncluttered view of human behavior.

```python
active_features = [
    'hour_sin',
    'hour_cos',
    #'day_sin',
    #'day_cos',
    'month_sin',
    'month_cos',
    'Humidity',
    'Apparent_Temp_C',
    'DayOfWeek',
    'Precipitation_mm',
    'Cloud_Cover',
    #'Is_Weekend',
    'Demand_Lag_24h',          
    #'Demand_ewm_24h',
    #'Temp_ewm_24h',            
    'Access to electricity (% of population)',
    'GDP (constant 2015 US$)'
]

columns_to_keep = active_features + ['DateTime', 'Target_Delta', 'Demand (MW)', 'Target_Next_Hour']
df_sandbox = df_master[columns_to_keep].copy()


train_mask = df_sandbox['DateTime'].dt.year <= 2023
test_mask = (df_sandbox['DateTime'] >= '2024-01-01 00:00:00') & \
            (df_sandbox['DateTime'] <= '2025-06-17 11:00:00')


X_train_box = df_sandbox.loc[train_mask, active_features]
y_train_box = df_sandbox.loc[train_mask, 'Target_Delta']

X_test_box = df_sandbox.loc[test_mask, active_features]
y_test_box = df_sandbox.loc[test_mask, 'Target_Delta']

current_demand_test = df_sandbox.loc[test_mask, 'Demand (MW)']
actual_future_demand = df_sandbox.loc[test_mask, 'Target_Next_Hour']

sandbox_model = xgb.XGBRegressor(n_estimators=500, learning_rate=0.05, max_depth=6, random_state=42, n_jobs=-1)
sandbox_model.fit(X_train_box, y_train_box)

predicted_deltas = sandbox_model.predict(X_test_box)
final_predictions = current_demand_test + predicted_deltas

sandbox_mape = mean_absolute_percentage_error(actual_future_demand, final_predictions)
print(f"MAPE: {sandbox_mape:.4%}\n")

df_sandbox_imp = pd.DataFrame({
    'Feature': active_features,
    'Importance': sandbox_model.feature_importances_
}).sort_values(by='Importance', ascending=False).reset_index(drop=True)

plt.figure(figsize=(12, 6))
sns.barplot(x='Importance', y='Feature', data=df_sandbox_imp, palette='mako', hue='Feature', legend=False)
plt.title(f'Feature Importance ({len(active_features)} Features)', fontsize=16, fontweight='bold')
plt.xlabel('Relative Importance', fontsize=12)
plt.ylabel('Feature', fontsize=12)
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```
